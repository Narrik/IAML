## Description of the dataset

In this part of the assignment we will explore the MNIST digits dataset. We expect the digits to lie in a lower-dimensional manifold and want to examine the representation we get by applying Principal Components Analysis (PCA). PCA maps the data into a new space by effectively rotating the base vectors of the input space to the directions with the highest variance.

### Description of the Data

This MNIST Dataset is a collection of handwritten digits. The samples are partitioned (nearly) evenly across the 10 different digit classes {0, 1, . . . , 9}. We use a preprocessed version for which the data are $8 \times 8$ pixel images containing one digit each. For further details on how the digits are preprocessed, see the sklearn documentation. The images are grayscale, with each pixel taking values in {0, 1, . . . , 16}, where 0 corresponds to black (weakest intensity) and 16 corresponds to white (strongest intensity). Therefore, the dataset is a N × 64
dimensional matrix where each dimension corresponds to a pixel from the image and N is the number of
images.

Set `svd_solver='full'` and keep all other parameters at default values. Then plot the *percentage* variance explained by each dimension as a `bar`-graph, in descending order of variance. *Hint: see the `explained_variance_ratio_` attribute of the PCA object.* On the same graph, plot (line) the cumululative explained variance as you add components (again in descending order of variance explained). Indicate the point at which 80% of the variance is explained (e.g. using matplotlib's [`vlines()`](https://matplotlib.org/2.2.3/api/_as_gen/matplotlib.pyplot.vlines.html)): label the number of components at this point (e.g. in the Legend). Make sure to distinguish between the plots and label them appropriately.<br>&nbsp;&nbsp;*Tip: you may need to use [`twinx()`](https://matplotlib.org/2.2.3/gallery/api/two_scales.html#sphx-glr-gallery-api-two-scales-py) to better combine the two plots. Also, for consistency, number components starting from 0.* <br>
&nbsp;&nbsp;**(b)** [Text] By commenting on the shape of the plot, why do you think 80% of the variance seems like a reasonable threshold? *Hint: Think about why we do PCA*. (<3 sentences)<br>
&nbsp;&nbsp;**(c)** [Text] What can you say about the number of components required to explain 80% of the variance? (<4 sentences)

````python
#(a) # Your Code goes here:
# Calculate PCA components
pca = PCA(n_components=digits.data.shape[1], svd_solver='full')
pca.fit(digits.data)
# Plot required information
fig, ax = plt.subplots(figsize=(16,6))
# Plot first the bargraph
plt.bar(x=range(64), height=pca.explained_variance_ratio_, label ='explained variance ratio')
ax.set_ylabel('explained variance ratio',color = 'blue')
ax.tick_params(axis='y', labelcolor='blue')
plt.xticks(range(64))
# Plot the cumulative line
ax2 = ax.twinx()
cumsum_evr =cumsum(pca.explained_variance_ratio_)
ax2.plot(cumsum_evr,color = 'red', label='cumulative explained variance ratio')
ax2.set_ylabel('cumulative explained variance ratio',color = 'red')
ax2.tick_params(axis='y', labelcolor='red')
# Plot the line of 80% variance
plt.vlines(x=np.argwhere(cumsum_evr > 0.8)[0], ymin=0, ymax=0.8, colors='black', label='80% of variance at {} components'.format(np.argwhere(cumsum_evr > 0.8)[0]))

lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=5)
plt.show()
````

(b) ***Your answer goes here:***

In our graph we see that past 4 components is where "the mountain stops and the rubble begins" in terms of gaining information, so one option could be only consider these 4 components, however that would only account for about 50% of the variance in the data, which means we are essentially losing half the information, on the other end of the spectrum, if we wanted to account for 95% of the variance we would need 28 components, which is almost a half of what we started with. 80% here seems like a good middle ground for not losing too much information and still greatly reducing the amount of dimensions (to 12).

(c) ***Your answer goes here:***

Our goal is to reduce the dimensionality as much as possibly without losing too much information in the process. Considering our data was 64-dimensional at the start, requiring 12 components to cover 80% of the variance means performing PCA was meaningful. Also it seems very reasonable, as from the heatmaps before we could see that it was mostly the same set of pixels in the middle that had the highest amount of variance.

To do this, we will first project the data onto the PCA space, and then reconstruct it using successively more dimensions (eigen-vectors). To save us time, we will do this for the first image (`digits.data[0,:]`) which should itself be a 0.

Using the PCA object's `transform` method, project this into the PCA space, and then reconstruct the original digit by multiplying each dimension with the projection value. Show the reconstructed digit, and also print the Mean Absolute Differences (MAD) between the two. *Tip: the MAD should be 0 - if it is not, you may be forgetting to add something...* Label everything appropriately.<br>

&nbsp;&nbsp;**(b)** [Code] Now we can analyse the quality of the reconstruction as we include more components. For the same digit, plot the reconstruction with `[0, 1, 2, 5, 15, 30]` components. For each reconstruction, compute and print (within the plot title) the MAD between the reconstruction and the original value.<br>

````python
#(a) # Your Code goes here:
zero = digits.data[0,:]
fig, ax = plt.subplots(1,2, figsize =  (15,5))
plt.subplot(1,2,1)
sns.heatmap(zero.reshape(8,8), cmap='binary', vmin = 0)
plt.title('Original Digit')
plt.xlabel('Column pixels')
plt.ylabel('Row pixels')
zero_pca = pca.transform(zero.reshape(1,-1))
res_zero = np.dot(zero_pca,pca.components_)
res_zero += pca.mean_
plt.subplot(1,2,2)
sns.heatmap(res_zero.reshape(8,8), cmap='binary', vmin = 0)
plt.title('Reconstructed digit using PCA')
plt.xlabel('Column pixels')
plt.ylabel('Row pixels')
print('Mean Absolute Difference {:.5f}'.format(np.mean(np.abs(res_zero-zero))))
plt.show()
````

````python
#(b) # Your Code goes here:
components = [0, 1, 2, 5, 15, 30]
fig, ax = plt.subplots(3,2, figsize =  (20,20))

for i,n in enumerate(components):
    res_zero = np.dot(zero_pca[:,:n],pca.components_[:n,:])
    res_zero += pca.mean_
    plt.subplot(3,2,i+1)
    sns.heatmap(res_zero.reshape(8,8), cmap='binary', vmin = 0)
    plt.title('Components: {}, MAD: {:.5f}'.format(n,np.mean(np.abs(res_zero-zero))))
    plt.xlabel('Column pixels')
    plt.ylabel('Row pixels')

plt.show()
````

(c) ***Your answer goes here:***

In our example we are working with digit '0', from **Q2.2** we saw that the 2nd principal component is responsible for the "U" shape in an image so we should be able to distinguish '0' after adding 2 principal components to the data. Looking at our reconstruction we can say that this is the case, as the MAD is lowered significantly after adding the 2nd PC and the digit becomes recogniseable. Naturally as we keep adding more principal components the image becomes clearer and clearer and MAD gets closer and closer to 0, however we were able to recognize the class of the digit just after 2 PC which means we applied PCA successfully, and for recognising '0' we went down from using 64 dimensions to just 2 dimensions (note that for a different digit such as '7' we would need more principal components to be able to recognise it).
